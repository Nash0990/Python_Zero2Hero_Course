```python
import random

#secretNum = random.randint(1,100)
secretNum = 50
print("Welcome to the number guessing game.\nYou have to guess the number being generated by our random number generator between 1 and 100.\nPlease enter \"Exit\" if you want to quit!\n\n")

guessList = []
count = 0
guess = 0

num = input("Please enter your first guess: ")

while guess == 0:
        if num.isdigit():
            guessList.append(int(num))
            
            if guessList[count] < 1 or guessList[count] > 100:
                
                print(f'{guessList[count]} is out of range.')
                guessList.append(int(input('Please guess another number: ')))
            
            elif guessList[count] == secretNum:
                print(f'Congratulations, you have guessed the correct number {secretNum} in {count} guesses.')
                guess = 1
                
            elif count == 0 and (guessList[count] < (secretNum + 10) and guessList[count] > (secretNum - 10) ):
                print('WARM!!!!!')
                guessList.append(int(input('Please guess another number: ')))
            
            elif count == 0 and (guessList[count] > (secretNum + 10) or guessList[count] < (secretNum - 10) ):
                print('COLD!!!!!')
                guessList.append(int(input('Please guess another number: ')))
            
            elif count > 0 and abs(guessList[-1] - secretNum) > abs(guessList[-2] - secretNum):
                print('Warmer!!!')
                guessList.append(int(input('Please guess another number: ')))
                
            else:
                print('Colder!!!')
                guessList.append(int(input('Please guess another number: ')))
            print(count)
            count += 1
        
        else:
            print(f'{guessList[count]} is not a number, please enter number between 1 and 100.')
```

```python

import random
secretNum = random.randint(1,100)

print("Welcome to the number guessing game.\nYou have to guess the number being generated by our random number generator between 1 and 100.\nPlease enter \"Exit\" if you want to quit!\n\n")
guessList = []
guess = 0
num = int(input("Please enter your first guess: "))
guessList.append(num)
while guess == 0:
    if guessList[-1] < 0 or guessList[-1] > 100:
        print(f"{guessList[-1]} is out of bounds.")
        guessList.append(int(input('Please entre a number between 1 and 100:')))
    elif guessList[-1] == secretNum:
        print(f'Congratulations, you have guessed the correct number {secretNum} in {len(guessList)} guesses.')
        guess = 1
    elif len(guessList) == 1 and (guessList[-1] > secretNum - 10 and guessList[-1] < secretNum + 10):
        print('Warm!!!!')
        guessList.append(int(input('Please entre a number between 1 and 100:')))
    elif len(guessList) == 1 and (guessList[-1] < secretNum - 10 or guessList[-1] > secretNum + 10):
        print('Cold!!!!')
        guessList.append(int(input('Please entre a number between 1 and 100:')))
    elif len(guessList) > 1 and abs(secretNum - guessList[-2]) > abs(secretNum - guessList[-1]):
        print('Warmer!!!!')
        guessList.append(int(input('Please entre a number between 1 and 100:')))
    elif len(guessList) > 1 and abs(secretNum - guessList[-2]) < abs(secretNum - guessList[-1]):
        print('Colder!!!!')
        guessList.append(int(input('Please entre a number between 1 and 100:')))
        
```
